```python
import pandas as pd
import numpy as np
import csv
asia_dat = pd.read_csv("asia.csv")
euro_dat = pd.read_csv("euro.csv")
gcc_dat = pd.read_csv("gcc.csv")
```

```python
euro_dat.head()
```

```python
euro_dat.loc[euro_dat["class"] == "4star", "class"] = "4 star"
euro_dat.head()
```

```python
euro_dat["value"] = euro_dat["value"].str.replace("-", "0")
euro_dat["value"] = pd.to_numeric(euro_dat["value"])

asia_dat["value"] = asia_dat["value"].str.replace("-", "0")
asia_dat["value"] = pd.to_numeric(asia_dat["value"])

gcc_dat["guests"] = gcc_dat["guests"].str.replace("-", "0")
gcc_dat["guests"] = pd.to_numeric(gcc_dat["guests"])
gcc_dat["nights"] = gcc_dat["nights"].str.replace("-", "0")
gcc_dat["nights"] = pd.to_numeric(gcc_dat["nights"])
```

```python
print(euro_dat["value"].dtype)
print(asia_dat["value"].dtype)
print(gcc_dat["guests"].dtype)
print(gcc_dat["nights"].dtype)
```

```python
euro_dat.loc[
    (euro_dat["year"] == 1998) &
    (euro_dat["class"] == "5 star") &
    (euro_dat["emirate"] == "abu dhabi") &
    (euro_dat["nationality"] == "Britan")
]
```

```python
# euro_dat = euro_dat.drop_duplicates(["year", "class", "item", 
#                                      "emirate", "nationality"], keep="first")

euro_dat = euro_dat.groupby(["year", "class", "emirate", 
                             "nationality", "item"]).agg({
                                                "value": sum})
euro_dat.columns.name = None
euro_dat = euro_dat.reset_index()
euro_dat.head()
```

```python
euro_dat.loc[
    (euro_dat["year"] == 1998) &
    (euro_dat["class"] == "5 star") &
    (euro_dat["emirate"] == "abu dhabi") &
    (euro_dat["nationality"] == "Britan")
]
```

```python
euro_unstack = euro_dat.set_index(["year", "class", 
                             "emirate", "nationality", 
                             "item"])["value"].unstack(fill_value=0)
euro_unstack.columns.name = None
euro_unstack = euro_unstack.reset_index()
euro_unstack.loc[euro_unstack["year"] == 2000].head()
```

```python
asia_unstack = asia_dat.set_index(["year", "class", 
                             "emirate", "nationality", 
                             "item"])["value"].unstack(fill_value=0)
asia_unstack.columns.name = None
asia_unstack = asia_unstack.reset_index()

asia_unstack.loc[asia_unstack["year"] == 2000].head()
```

```python
gcc_dat.head()
```

```python
dat_merge = pd.concat([gcc_dat, asia_unstack, euro_unstack])
```

```python
dat_merge.loc[dat_merge["nationality"] == "Britan"].head()
#euro_unstack.loc[euro_unstack["nationality"] == "Britan"]
```

```python
dat_merge.emirate = dat_merge.emirate.str.strip()
dat_merge.loc[dat_merge["emirate"] == "dubai", "emirate"] = "Dubai"
dat_merge.loc[dat_merge["emirate"] == "abu dhabi", "emirate"] = "Abu Dhabi"
dat_merge.loc[dat_merge["emirate"] == "Ras Al Khaima & Fujeira", 
              "emirate"] = "Ras Al-Khaimah and Fujairah"
dat_merge.loc[dat_merge["emirate"] == "sharjah and Ajman", "emirate"] = "Ajman and Sharjah"
dat_merge.loc[dat_merge["emirate"] == "sharjah and ajman", "emirate"] = "Ajman and Sharjah"
dat_merge.loc[dat_merge["emirate"] == "Ajman", "emirate"] = "Ajman and Sharjah"
dat_merge.loc[dat_merge["emirate"] == "sharjah", "emirate"] = "Ajman and Sharjah"

dat_merge.loc[dat_merge["emirate"] == "Umm Al-Qiwain", "emirate"] = "Umm Al Quwain"
dat_merge.loc[dat_merge["emirate"] == "Ajman and Umm Al-Qiwain", "emirate"] = "Umm Al Quwain"

dat_merge.loc[dat_merge["class"] == "4star", "class"] = "4 star"
dat_merge.loc[dat_merge["class"] == "3 star ", "class"] = "3 star"
```

```python
dat_merge = dat_merge[dat_merge.year != 1885]
```

```python
def addMissingVal(dat):
    year = list(range(min(dat["year"]), max(dat["year"])))
    clas = list(set(dat["class"]))
    emirate = list(set(dat["emirate"]))
    nationality = list(set(dat["nationality"]))
    
    year_lt = len(year)
    class_lt = len(clas)
    emirate_lt = len(emirate)
    nation_lt = len(nationality)
    
    nation_cnt = year_lt * class_lt * emirate_lt 
    npnation = nationality * nation_cnt
    
    class_cnt = year_lt * emirate_lt * nation_lt
    npclass = clas * class_cnt
    
    emirate_cnt = year_lt * class_lt * nation_lt
    npemirate = emirate * emirate_cnt
    
    year_cnt = class_lt * emirate_lt * nation_lt
    npyear = year * year_cnt
    
    new = pd.DataFrame({"nationality": npnation, "emirate": npemirate, 
                        "year": npyear, "class": npclass, "guests": 1000, "nights": 1})
    
    combineMe = new[["year", "class", "nationality", "emirate", "guests", "nights"]]
    print(combineMe)
    
#     print(combineMe.loc[
#     (combineMe["year"] == 1989) &
#     (combineMe["class"] == "5 star") &
#     (combineMe.emirate == "Dubai")
#              ])
    new = pd.concat([dat, new])
    return new
```

```python
dat_merge = addMissingVal(dat_merge)
# print(dat_merge.loc[
#     (dat_merge.year == 1999) &
#     (dat_merge["class"] == "5 star") &
#     (dat_merge.emirate == "Dubai")
#              ])

```
